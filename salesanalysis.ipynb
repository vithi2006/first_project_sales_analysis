```python
import pandas as pd

# Load the CSV file
df = pd.read_csv("sales_data.csv")
print(df.head())
```

```python
print(df.isnull().sum())
#data cleaning. if all is zero its good
```

```python
#add total_sales column

# Total Sales = Quantity Ã— Unit Price
df['Total Sales'] = df['Quantity'] * df['Unit Price']
print(df)
```

```python
#using groupby()

# Sales by Region
sales_by_region = df.groupby('Region')['Total Sales'].sum()
print(sales_by_region)

# Sales by Product
sales_by_product = df.groupby('Product')['Total Sales'].sum()
print(sales_by_product)
```

```python
#visualize

import matplotlib.pyplot as plt

# Bar chart for Region
sales_by_region.plot(kind='bar', title='Total Sales by Region', ylabel='Sales ($)', xlabel='Region')
plt.show()

# Bar chart for Product
sales_by_product.plot(kind='bar', title='Total Sales by Product', ylabel='Sales ($)', xlabel='Product', color='orange')
plt.show()
```
